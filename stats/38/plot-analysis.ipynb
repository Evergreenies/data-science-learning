```python
batsman_season = pd.read_csv("./batsman_season_record.csv")

batsman_season
```

```python
plt.bar(batsman_season['batsman'], batsman_season['2015'], width=0.5)
```

```python
plt.bar(np.arange(batsman_season.shape[0]) - 0.2, batsman_season['2015'], width=0.2)
plt.bar(np.arange(batsman_season.shape[0]), batsman_season['2016'], width=0.2)
plt.bar(np.arange(batsman_season.shape[0]) + 0.2, batsman_season['2017'], width=0.2)

plt.xticks(np.arange(batsman_season.shape[0]), batsman_season['batsman'])
plt.show()
```

```python
plt.bar(batsman_season['batsman'], batsman_season['2015'], label='2015')
plt.bar(batsman_season['batsman'], batsman_season['2016'], bottom=batsman_season['2015'], label='2016')
plt.bar(batsman_season['batsman'], batsman_season['2017'], bottom=batsman_season['2016'] + batsman_season['2015'], label='2017')

plt.legend()
plt.show()
```
